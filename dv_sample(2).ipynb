```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


#IMPORTING CSV FILE
df=pd.read_csv('Mall_Customers.csv')
print(df)
```

```python
#ADDING DATE

df["date"] = pd.date_range(start='01-01-2025', periods=len(df),freq='D')
print(df)

```

```python
#CHART NO:1 SCATTER PLOT(Age vs. Spending Score):
plt.scatter(df['Age'],df['Spending Score (1-100)'],marker='*',color="pink")
plt.xlabel('AGE')
plt.ylabel('SPENDING SCORE')
plt.title("SCATTER PLOT OF AGE VS SPENDING SCORE ")
```

```python
#CHART NO:2 BAR PLOT (Gender Distribution):

gender_counts = df['Gender'].value_counts()
plt.barh(gender_counts.index, gender_counts.values, color=["pink","skyblue"])
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.title("GENDER DISTRIBUTION")
```

```python
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=["pink","skyblue"])
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.title("GENDER DISTRIBUTION")
```

```python
#CHART NO:3 LINE CHART (Date vs. Spending Score):

plt.plot(df['date'],df['Spending Score (1-100)'])
plt.xlabel("Date")
plt.ylabel("Spending Score (1-100)")
plt.show
```

```python
#CHART NO:4 PIE CHART (Gender Distribution):

plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```
