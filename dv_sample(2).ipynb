```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("Mall_Customers.csv")
print(df)
df.head()
df.describe()
```

```python
import seaborn as sns
sns.boxplot(data=df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']],palette= ['yellow', 'red','pink'])

plt.title('Box Plot')
plt.show()
```

```python
df.info()
```

```python
df['Annual Income (k$)'].unique()
```

```python
import seaborn as sns
sns.violinplot(x= 'Gender',y='Spending Score (1-100)', hue = 'Gender',data=df,split=True,palette=['lightblue','pink'])
plt.show()
```

```python
#count plot
plt.figure(figsize=(5,5))
sns.countplot(x='Gender',data=df  ,hue = "Gender",palette='Spectral')
plt.show()
```

```python
df['Annual Income (k$)'] = df['Annual Income (k$)'].str.replace('$', '', regex=False).str.strip()
df['Annual Income (k$)'].unique()
```

```python
sns.histplot(df['Spending Score (1-100)'],bins=10)
plt.title("HISTOGRAM")
plt.show
```

```python
sns.kdeplot(data=df['Spending Score (1-100)'],fill= True )
plt.title("KDE PLOT")
plt.show()
```

   1.  Data Loading and Initial Glance:

  2.  Data Profiling (ydata-profiling):

   3. Data Types and Cleaning:
        df.info() revealed that Annual Income (k$) is of object (string) type, which is incorrect for a numerical column. This was further confirmed by df['Annual Income (k$)'].unique() which showed values like '$20 '.
        Result: The '$' symbol and leading/trailing spaces were removed from the Annual Income (k$) column, making it ready for conversion to a numerical type (though the conversion itself wasn't shown in the last executed cells, this is the logical next step).
