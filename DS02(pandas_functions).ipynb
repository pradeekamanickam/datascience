# 3) HANDLING MISSING DATA'S

**df.isnull()**:        identifies missing values

**df.fillna(value)**: Fills missing values

**df.dropna()**: removes missing values/ spaces

```python
import pandas as pd


data = {'name ': ['aryan','harsh','pradee','priya'],'age': [None,None,15, None],'salary':[2000,5000, None, None]}
df = pd.DataFrame(data)
print(df)


missing_values=df.isnull()
print(missing_values)

rem_mvalues=df.dropna()
print(rem_mvalues)

fill_mvalues =df.fillna("5000")
print(fill_mvalues)
```

#4) DATA CLEANING

**df.drop_duplicates()**: removes duplicate rows

**df.rename(columns={})**: rename column

**df.astype(dtype)**: convert data types of colmns, by inserting the datatype we wanted to change.

**df.replae(old,new): replace values


```python
import pandas as pd

data = {'name ': ['aryan','harsh','pradee','priya'],'age': [15,None,15, None],'salary':[2000,5000, None, None]}
df = pd.DataFrame(data)
print(df)
drop=df.drop_duplicates("age")
print(drop)
new_col=df.rename(columns={})
print(new_col)
d_type = df.astype(bool)
print(d_type)
```

#   6)GROUPING AND AGGREGATION

**df.groupby()**: groups data by a column.

**df.agg()**: apply aggregation functions (sum(),mean())

**df.pivot_table()**:create pivot tables( splits the main table into parts)

```python
import pandas as pd
data = {'name ': ['aryan','harsh','pradee','priya'],'age': [5,None,15, None],'salary':[2000,5000, None, None]}
df = pd.DataFrame(data)
print(df)

group= df.groupby(by='name ')
print(group)

aggregrate=df.agg(func='salary').sum()
print(aggregrate)

split_table=df.pivot_table(index='name ', values='salary', aggfunc='sum')
print(split_table)
```
