```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#IMPORTING CSV FILE
df=pd.read_csv('Mall_Customers.csv')
print(df)
```

```python
#ADDING DATE COLOUMN
df["date"] = pd.date_range(start='01-01-2025', periods=len(df),freq='D')
print(df)

```

```python
#CHART NO:1 SCATTER PLOT

plt.scatter(df['Age'],df['Spending Score (1-100)'],marker='*',color="pink")
plt.xlabel('AGE')
plt.ylabel('SPENDING SCORE')
plt.title("SCATTER PLOT OF AGE VS SPENDING SCORE ")
```

```python
#CHART NO:2 BAR PLOT

gender_counts = df['Gender'].value_counts()
plt.barh(gender_counts.index, gender_counts.values, color=["pink","skyblue"])
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.title("GENDER DISTRIBUTION")
```

```python
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=["pink","skyblue"])
plt.xlabel('GENDER')
plt.ylabel('COUNT')
plt.title("GENDER DISTRIBUTION")
```
