## **Code :**





```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("tested.xls")
print(df.head())
```

```
df.describe()
```

```
#1. Box Plot of Age and Fare

plt.figure(figsize=(8,5))

sns.boxplot(
data=df[['Age','Fare']],
palette=["skyblue","pink"]
)

plt.title("Box Plot of Age and Fare")
plt.ylabel("Values")
plt.show()
```

## **INTERPRETATION**


The box plot shows that both Age and Fare contain outliers. Most passengers belong to a similar age group and paid relatively low fares, while a few passengers paid exceptionally high ticket fares.

```
#2.Violin Plot of Age by Survival

plt.figure(figsize=(8,5))

sns.violinplot(
data=df,
x="Survived",
y="Age",
palette=["skyblue","pink"]
)

plt.title("Age Distribution by Survival")
plt.show()
```

## **INTERPRETATION**

The violin plot shows that passengers of various age groups survived. Most passengers were adults, while children also had a noticeable survival rate. The distribution helps compare the age spread between survivors and non-survivors.


```
# Count Plot of Passenger Class

plt.figure(figsize=(7,5))

sns.countplot(data=df,x="Pclass",hue="Pclass",palette=["pink","skyblue","violet"])
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

## **INTERPRETATION**


The count plot shows that Third Class had the highest number of passengers, followed by First Class, while Second Class had comparatively fewer passengers.


```
#Bar Plot of Average Fare by Passenger Class

plt.figure(figsize=(7,5))

sns.barplot(data=df,x="Pclass",y="Fare",hue="Pclass",palette=["pink","skyblue","violet"])
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

plt.show()
```

## **INTERPRETATION**


The bar plot indicates that First Class passengers paid the highest average fare, while Third Class passengers paid the lowest fare. Ticket prices clearly vary according to passenger class.



```
#5. KDE Plot of Age Distribution

plt.figure(figsize=(7,5))

sns.kdeplot(data=df,x="Age",fill=True,color="pink")
plt.title("KDE Plot of Age Distribution")
plt.xlabel("Age")
plt.ylabel("Density")

plt.show()
```

## **INTERPRETATION**




```
#6.Scatter Plot of Age vs Fare by Survival

plt.figure(figsize=(8,5))

sns.scatterplot(data=df,x="Age",y="Fare",hue="Survived")
plt.title("Age vs Fare by Survival")
plt.show()
```

## **INTERPRETATION**

The KDE plot shows that most passengers were between 20 and 40 years of age. Very young and elderly passengers were fewer in number, resulting in lower density at both ends of the graph.
