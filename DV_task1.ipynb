## ***CODE :***





```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


#IMPORTING CSV FILE
df = pd.read_csv("tested.xls")
print(df.head(5))
```

```
print(df.info())
```

```
#CHART NO:1 SCATTER PLOT – Age vs Fare:

plt.figure(figsize=(6,6))
plt.scatter(df['Age'], df['Fare'],color=['pink'],marker='*')
plt.title("SCATTER PLOT – Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

## ***INTERPRETATION:***



*   No strong relationship exists between age and fare.
*  Higher fares are spread across different age groups.
*  A few passengers paid exceptionally high fares, indicating outliers.
*  Scatter plots help identify relationships and unusual observations.








```
#CHART NO:2 LINE CHART - Average Fare by Passenger Class:

avg_fare = df.groupby('Pclass')['Fare'].mean()

plt.figure(figsize=(5,5))
plt.plot(avg_fare.index, avg_fare.values, marker='*',color=('lightblue'))
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.xticks(avg_fare.index)
plt.show()
```

## ***INTERPRETATION:***


* Average fare decreases from first class to third class.
* First-class passengers paid the highest fares.
* The line chart clearly shows the trend across passenger classes.
* It indicates a relationship between travel class and ticket fare.







```
#CHART NO:3 PIE CHART – Survival Percentage:

survival = df['Survived'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(survival.values,labels=['Not Survived','Survived'],autopct='%1.1f%%',colors=['pink','lightblue'])
plt.title("PIE CHART – Survival Percentage")
plt.show()
```

## ***INTERPRETATION:***

* The chart shows the proportion of passengers who survived and those who did not.
* The majority of passengers did not survive.
* It provides an easy comparison of survival outcomes.
* Pie charts are useful for displaying percentages of a whole.







```
#CHART NO: 4 Histogram – Age Distribution

sns.histplot(df["Age"], bins=20, kde=True,color='pink',edgecolor='pink')
plt.title("Histogram – Age Distribution",color='violet')
plt.show()
```

## ***INTERPRETATION:***

* Most passengers were between 20 and 40 years of age.
* There were relatively fewer elderly passengers.
* The age distribution is slightly right-skewed.
* This chart helps understand how passenger ages are distributed.







```
#CHART NO : 5 BOX Plot – Fare Distribution
plt.figure(figsize=(8,5))
plt.boxplot(df['Fare'].dropna())
plt.title("BOX Plot - Fare Distribution",color='violet')
plt.ylabel("Fare")
plt.show()
```

## ***INTERPRETATION:***

* The median fare lies near the center of the box.
* Several high-fare outliers are visible.
* Most passengers paid relatively lower fares.
* Box plots are useful for detecting outliers and understanding data spread.
