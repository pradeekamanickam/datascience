```python
import pandas as pd

data={'BOOKS':['AI Basics' ,'Data Science', 'Python 101',' Algorithms', 'Machine Learning',]}
df= pd.DataFrame(data)
print(df)

#sorting
sorted=df.sort_values('BOOKS')
print(sorted)

#Inserting Cyber Security into 3rd position
new=df.replace('Python 101','Cybersecurity')
print(new)

# Adding a new book 'Chemistry'
df = pd.concat([df, pd.DataFrame({'BOOKS': ['Chemistry']})], ignore_index=3)
print('DataFrame after adding Chemistry:')
print(df)
```
