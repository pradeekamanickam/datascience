# 1)Data Preprocessing:
(filling the missing data's using fillna())

```python
import pandas as pd
data = {'name ': ['aryan','harsh','pradee','priya'],'age': [5,None,15, None],'salary':[2000,5000, None, None]}
df = pd.DataFrame(data)
df['age'].fillna(df['age'].mean(),inplace= True)
df['salary'].fillna(5,inplace=True)
print(df)
```

```python
#example(1)
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Score': [85, 99, 95]}
df = pd.DataFrame(data)
df.sort_values('Score', inplace=True)
new_df = df.sort_values('Score',inplace=False)
print(data)
print(df)
print(new_df)
```

# 2) Data cleaning
 (to drop duplicates rows)




```python
import pandas as pd
data = {'name ': ['aryan','harsh','pradee','priya'],'age': [15,15,18, None],'salary':[5000,5000, None, None]}
df=pd.DataFrame(data)
print(df)
df=df.drop_duplicates('age')
print(df)

#to rename in pandas, we use rename()
df.rename(columns={'salary':'salary_Amt'},inplace=True)
print(df)
```

# 3)Data transformation:

Filters and sorts data for better insights.

```python
import pandas as pd
df=pd.DataFrame({'PRODUCTS':['LAPTOP','PHONE','HOODIE','JEANS'],'CATEGORY':['ELECTRONICS','CLOTHING','ELECTRONICS','CLOTHING'],'PRICE':[50000,5000,500,800]} )
print(df)

filtered_df= df.query("CATEGORY == 'ELECTRONICS'")    # 'query()' filters the 'category' column of having vales "ELECTRONICS"
sorted_df=filtered_df.sort_values(by='PRICE')         # 'sort_values()'csorts the data according to 'PRICE' column.
print(sorted_df)
```

# 4) Data Visualisation:

Represent data grafically for insights,like plotting data using matplotlib.

```python
from matplotlib.colors import BASE_COLORS
from matplotlib.typing import ColourType
import matplotlib.pyplot as plt
import pandas as pd

df=pd.DataFrame({'PRODUCTS':['LAPTOP','PHONE','HOODIE','JEANS'],'CATEGORY':['ELECTRONICS','CLOTHING','ELECTRONICS','CLOTHING'],'PRICE':[50000,5000,500,800]} )
plt.bar(df['CATEGORY'],df['PRICE'],color=['green','orange','white','red'])
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```
